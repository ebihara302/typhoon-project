帆の本数30本に対するMCHcapa変化の感度解析

```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np
import polars as pl

for e_i in range(2):
    e_PG = 0.4 + 0.1*e_i

    #データの読み込み
    data_MCH = pl.read_csv("output_Folder/TPGshipdata/change_e_PG_LiftGenerator/TPGship_model_and_result_MCH_ePG=" + str(e_PG) + "_and_Liftgene.csv")
    data_Loss = pl.read_csv("output_Folder/TPGshipdata/change_e_PG_LiftGenerator/TPGship_model_and_result_Losselect_ePG=" + str(e_PG) + "_and_Liftgene.csv")

    # e_PG = 0.4の時のデータを抽出
    data_MCH = data_MCH.filter(data_MCH['e_PG'] == e_PG)
    data_Loss = data_Loss.filter(data_Loss['e_PG'] == e_PG)


    # データのcsvファイルを出力
    data_MCH.write_csv("output_Folder/TPGshipdata/change_e_PG_LiftGenerator/TPGship_model_and_result_MCH_ePG=" + str(e_PG) + "_and_Liftgene.csv")
    data_Loss.write_csv("output_Folder/TPGshipdata/change_e_PG_LiftGenerator/TPGship_model_and_result_Losselect_ePG=" + str(e_PG) + "_and_Liftgene.csv")
```

### 台風発電船の発電機を格納できる場合に発電効率のみを向上させた場合のMCH生成量の変化

```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np
import polars as pl

for e_i in range(3):
    e_PG = 0.3 + 0.1*e_i

    #データの読み込み
    data = pl.read_csv("output_Folder/TPGshipdata/change_e_PG_LiftGenerator/TPGship_model_and_result_MCH_ePG=" + str(e_PG) + "_and_Liftgene.csv")


    #データの確認
    x = data["Storage[GWh]"]
   
    y_2017 = data["2017 Total Gene[GWh]"]
    y_2018 = data["2018 Total Gene[GWh]"]
    y_2019 = data["2019 Total Gene[GWh]"]
    y_2020 = data["2020 Total Gene[GWh]"]
    y_2021 = data["2021 Total Gene[GWh]"]
    y_2022 = data["2022 Total Gene[GWh]"]

    
    


    #年間平均のリスト
    y_ave = []
    for i in range(len(y_2017)):
        y_ave.append((y_2017[i] + y_2018[i] + y_2019[i] + y_2020[i] + y_2021[i] + y_2022[i])/6)


    #データのプロット
    #ここでスタイルを指定する
    #今回は'science'と'ieee'を利用してみる
    plt.style.use(['science','ieee','high-vis','grid'])


    #図の表示
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1) 
    
    ax.set_xlim(xmin=0,xmax=300) #x軸の範囲を指定
    ax.set_ylim(ymin=0,ymax=800) #y軸の範囲を指定
    ax.plot(x,y_2017,label = '2017') #2017年の発電量を表示
    ax.plot(x,y_2018,label = '2018') #2018年の発電量を表示
    ax.plot(x,y_2019,label = '2019') #2019年の発電量を表示
    ax.plot(x,y_2020,label = '2020') #2020年の発電量を表示
    ax.plot(x,y_2021,label = '2021') #2021年の発電量を表示
    ax.plot(x,y_2022,label = '2022') #2022年の発電量を表示
    #黒色の太い実線で年間平均の発電量を表示
    ax.plot(x,y_ave,color = 'black',linewidth = 2,label = 'Average')


    ax.set_title('e_PG: ' + str(e_PG))
    #ax.legend(title = 'Year',loc = 'lower right',fancybox = False, frameon = True) #凡例の細かい設定
    ax.legend(title = 'Year of Operation'  ,loc = 'upper right',fancybox = False, frameon = True, ncol = 3) #凡例の細かい設定
    ax.set(xlabel = 'Q_capacityMCH[GWh]') #x軸のラベル
    ax.set(ylabel = 'Annual MCH production[GWh]') #y軸のラベル
    plt.show() 
```

### 台風発電船の発電機を格納できる場合とできない場合の各発電効率における年平均MCH生成量を比較

```python
# 発電機格納タイプと非格納タイプのe_PG = 0.3~0.5の時の年平均MCH生成量をそれぞれのリストに格納
y_ave_genelift_e03 = []
y_ave_genelift_e04 = []
y_ave_genelift_e05 = []

y_ave_nolift_e03 = []
y_ave_nolift_e04 = []
y_ave_nolift_e05 = []

e_PG = 0.3
data_genelift = pl.read_csv("output_Folder/TPGshipdata/change_e_PG_LiftGenerator/TPGship_model_and_result_MCH_ePG=" + str(e_PG) + "_and_Liftgene.csv")
x = data_genelift["Storage[GWh]"]

for e_i in range(3):
    e_PG = 0.3 + 0.1*e_i

    #データの読み込み
    data_genelift = pl.read_csv("output_Folder/TPGshipdata/change_e_PG_LiftGenerator/TPGship_model_and_result_MCH_ePG=" + str(e_PG) + "_and_Liftgene.csv")
    data_nolift = pl.read_csv("output_Folder/TPGshipdata/change_e_PG/TPGship_model_and_result_MCH_ePG=" + str(e_PG) + ".csv")

    # 2017年から2022年までの年平均のMCH生成量を計算
    # 発電機格納タイプの場合
    y_2017_genelift = data_genelift["2017 Total Gene[GWh]"]
    y_2018_genelift = data_genelift["2018 Total Gene[GWh]"]
    y_2019_genelift = data_genelift["2019 Total Gene[GWh]"]
    y_2020_genelift = data_genelift["2020 Total Gene[GWh]"]
    y_2021_genelift = data_genelift["2021 Total Gene[GWh]"]
    y_2022_genelift = data_genelift["2022 Total Gene[GWh]"]

    # 発電機非格納タイプの場合
    y_2017_nolift = data_nolift["2017 Total Gene[GWh]"]
    y_2018_nolift = data_nolift["2018 Total Gene[GWh]"]
    y_2019_nolift = data_nolift["2019 Total Gene[GWh]"]
    y_2020_nolift = data_nolift["2020 Total Gene[GWh]"]
    y_2021_nolift = data_nolift["2021 Total Gene[GWh]"]
    y_2022_nolift = data_nolift["2022 Total Gene[GWh]"]    

    # e_PGの数値で場合わけしてリストに各Storageの数値での年平均のMCH生成量を格納
    if e_PG == 0.3:
        for i in range(len(y_2017_genelift)):
            y_ave_genelift_e03.append((y_2017_genelift[i] + y_2018_genelift[i] + y_2019_genelift[i] + y_2020_genelift[i] + y_2021_genelift[i] + y_2022_genelift[i])/6)
            y_ave_nolift_e03.append((y_2017_nolift[i] + y_2018_nolift[i] + y_2019_nolift[i] + y_2020_nolift[i] + y_2021_nolift[i] + y_2022_nolift[i])/6)
    elif e_PG == 0.4:
        for i in range(len(y_2017_genelift)):
            y_ave_genelift_e04.append((y_2017_genelift[i] + y_2018_genelift[i] + y_2019_genelift[i] + y_2020_genelift[i] + y_2021_genelift[i] + y_2022_genelift[i])/6)
            y_ave_nolift_e04.append((y_2017_nolift[i] + y_2018_nolift[i] + y_2019_nolift[i] + y_2020_nolift[i] + y_2021_nolift[i] + y_2022_nolift[i])/6)
    elif e_PG == 0.5:
        for i in range(len(y_2017_genelift)):
            y_ave_genelift_e05.append((y_2017_genelift[i] + y_2018_genelift[i] + y_2019_genelift[i] + y_2020_genelift[i] + y_2021_genelift[i] + y_2022_genelift[i])/6)
            y_ave_nolift_e05.append((y_2017_nolift[i] + y_2018_nolift[i] + y_2019_nolift[i] + y_2020_nolift[i] + y_2021_nolift[i] + y_2022_nolift[i])/6)
    
    
#データのプロット
#ここでスタイルを指定する
#今回は'science'と'ieee'を利用してみる
plt.style.use(['science','ieee','high-vis','grid'])

# 図の表示
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_xlim(xmin=0,xmax=300) #x軸の範囲を指定
ax.set_ylim(ymin=0,ymax=350) #y軸の範囲を指定
ax.plot(x,y_ave_genelift_e03,label = 'Storable Generators(e_PG=0.3)') #e_PG=0.3の発電機格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_nolift_e03,label = 'Non-Storable Generators(e_PG=0.3)') #e_PG=0.3の発電機非格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_genelift_e04,label = 'Storable Generators(e_PG=0.4)') #e_PG=0.4の発電機格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_nolift_e04,label = 'Non-Storable Generators(e_PG=0.4)') #e_PG=0.4の発電機非格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_genelift_e05,label = 'Storable Generators(e_PG=0.5)') #e_PG=0.5の発電機格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_nolift_e05,label = 'Non-Storable Generators(e_PG=0.5)') #e_PG=0.5の発電機非格納タイプの場合の年平均MCH生成量を表示
ax.set_title('Average Annual MCH production')
#凡例はサイズを小さめにして表示
ax.legend(title = 'Type of Generator',loc = 'lower right',fancybox = False, frameon = True, ncol = 2, fontsize = 4) #凡例の細かい設定

ax.set(xlabel = 'Q_capacityMCH[GWh]') #x軸のラベル
ax.set(ylabel = 'Annual MCH production[GWh]') #y軸のラベル
plt.show()



```
