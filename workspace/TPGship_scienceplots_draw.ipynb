蓄電容量と発電量の関係

```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np
import polars as pl

#データの読み込み
data = pl.read_csv('jasnaoe_storage_and_base/jasnaoe_result_2017-2022_Minamitori_capa_40%_monohull.csv',encoding="shift-jis")


#データの確認
storage = data["STORAGE[GWh]"]
x = []
for i in range(len(storage)):
    x.append(storage[i])
gene_2017 = data["2017 POWER GENERATION[GWh]"]
y_2017 = []
for i in range(len(gene_2017)):
    y_2017.append(gene_2017[i])
gene_2018 = data["2018 POWER GENERATION[GWh]"]
y_2018 = []
for i in range(len(gene_2018)):
    y_2018.append(gene_2018[i])
gene_2019 = data["2019 POWER GENERATION[GWh]"]
y_2019 = []
for i in range(len(gene_2019)):
    y_2019.append(gene_2019[i])
gene_2020 = data["2020 POWER GENERATION[GWh]"]
y_2020 = []
for i in range(len(gene_2020)):
    y_2020.append(gene_2020[i])
gene_2021 = data["2021 POWER GENERATION[GWh]"]
y_2021 = []
for i in range(len(gene_2021)):
    y_2021.append(gene_2021[i])
gene_2022 = data["2022 POWER GENERATION[GWh]"]
y_2022 = []
for i in range(len(gene_2022)):
    y_2022.append(gene_2022[i])

#データのプロット
#ここでスタイルを指定する
#今回は'science'と'ieee'を利用してみる
plt.style.use(['science','ieee','high-vis','grid'])


#図の表示
fig = plt.figure()
ax = fig.add_subplot(1,1,1) 
ax.set_xlim(xmin=0,xmax=200) #x軸の範囲を指定
ax.set_ylim(ymin=0,ymax=300) #y軸の範囲を指定
ax.plot(x,y_2017,label = '2017') #2017年の発電量を表示
ax.plot(x,y_2018,label = '2018') #2018年の発電量を表示
ax.plot(x,y_2019,label = '2019') #2019年の発電量を表示
ax.plot(x,y_2020,label = '2020') #2020年の発電量を表示
ax.plot(x,y_2021,label = '2021') #2021年の発電量を表示
ax.plot(x,y_2022,label = '2022') #2022年の発電量を表示

#ax.legend(title = 'Year',loc = 'lower right',fancybox = False, frameon = True) #凡例の細かい設定
ax.legend(title = 'Year of Operation' ,title_fontsize = 8 ,loc = 'lower right',fancybox = False, frameon = True, ncol = 2) #凡例の細かい設定
ax.set(xlabel = 'Power Storage Amount[GWh]') #x軸のラベル
ax.set(ylabel = 'Annual Power Generation[GWh]') #y軸のラベル
fig.show() 
```

待機位置の変更による影響調査

```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np
import polars as pl

#データの読み込み
data = pl.read_csv('jasnaoe_storage_and_base/jasnaoe_result_2018-2022_Minamitori_standbyposition_40%_monohull.csv',encoding="shift-jis")

#データの整理
year = data["YEAR"]
labels = []
for i in range(len(year)):
    labels.append(year[i])
x = data["BASE[GWh]"]
height_base = []
for i in range(len(x)):
    height_base.append(x[i])
x = data["E[GWh]"]
height_E = []
for i in range(len(x)):
    height_E.append(x[i])
x = data["W[GWh]"]
height_W = []
for i in range(len(x)):
    height_W.append(x[i])
x = data["S[GWh]"]
height_S = []
for i in range(len(x)):
    height_S.append(x[i])
x = data["N[GWh]"]
height_N = []
for i in range(len(x)):
    height_N.append(x[i])


plt.style.use(['science','ieee','high-vis','grid'])

left = np.arange(len(height_base))  # numpyで横軸を設定

width = 0.15
fig = plt.figure()
ax = fig.add_subplot(1,1,1) 


plt.bar(left, height_base, width=width, align='center', label='Base')
plt.bar(left+width, height_E, width=width, align='center', label='E')
plt.bar(left+width*2, height_W, width=width, align='center', label='W')
plt.bar(left+width*3, height_S, width=width, align='center', label='S')
plt.bar(left+width*4, height_N, width=width, align='center', label='N')

plt.xticks(left + width*2, labels)
ax.set(ylabel = 'Average Annual Power Generation[GWh]') #y軸のラベル
ax.legend(loc = 'lower center',fancybox = False, frameon = True, bbox_to_anchor=(0.5, -0.2), ncol = 5) #凡例の細かい設定
fig.show()
```

帆の本数に対する消費電力量のグラフ

```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np
import polars as pl

# データの読み込み
data = pl.read_csv('output_Folder/TPGshipdata/sailnum_change/Loss_sailnum_20~50.csv', encoding="shift-jis")

# 帆の本数
sail_num = data["Sail Num"]

# 各年の消費電力量
total_loss_2017 = data["2017 Total Loss[GWh]"]
total_loss_2018 = data["2018 Total Loss[GWh]"]
total_loss_2019 = data["2019 Total Loss[GWh]"]
total_loss_2020 = data["2020 Total Loss[GWh]"]
total_loss_2021 = data["2021 Total Loss[GWh]"]
total_loss_2022 = data["2022 Total Loss[GWh]"]

# データのプロット
plt.style.use(['science','ieee','high-vis','grid'])

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# x軸の範囲を指定
ax.set_xlim(xmin=0, xmax=max(sail_num))
ax.set_ylim(ymin=0, ymax=max(max(total_loss_2017), max(total_loss_2018), max(total_loss_2019), max(total_loss_2020), max(total_loss_2021), max(total_loss_2022)) + 10)

# 各年の消費電力量のプロット
ax.plot(sail_num, total_loss_2017, label='2017')
ax.plot(sail_num, total_loss_2018, label='2018')
ax.plot(sail_num, total_loss_2019, label='2019')
ax.plot(sail_num, total_loss_2020, label='2020')
ax.plot(sail_num, total_loss_2021, label='2021')
ax.plot(sail_num, total_loss_2022, label='2022')

# 凡例の設定
ax.legend(title='Year of Operation', title_fontsize=8, loc='lower right', fancybox=False, frameon=True, ncol=2)

# 軸ラベルの設定
ax.set(xlabel='Number of Sails')
ax.set(ylabel='Total Power Consumption[GWh]')

plt.show()
```
