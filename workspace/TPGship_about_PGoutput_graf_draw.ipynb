```python
#ライブラリインポート
import matplotlib.pyplot as plt
import scienceplots
import numpy as np
import polars as pl
```

```python
# storage 50GWh~300GWhの時の年平均MCH生成量をそれぞれのリストに格納
y_ave_MCH_sto50 = []
y_ave_MCH_sto100 = []
y_ave_MCH_sto150 = []
y_ave_MCH_sto200 = []
y_ave_MCH_sto250 = []
y_ave_MCH_sto300 = []

y_ave_loss_sto50 = []
y_ave_loss_sto100 = []
y_ave_loss_sto150 = []
y_ave_loss_sto200 = []
y_ave_loss_sto250 = []
y_ave_loss_sto300 = []


storage = 50

#データの読み込み
data = pl.read_csv("output_Folder/TPGshipdata/change_PGoutput/TPGship_model_and_result_MCH_capaMCH=" + str(storage) + "GWh.csv")
x = data["PG output[GWh]"]
x2 = data["PG radious[m]"]


for sto_i in range(5):
    storage = 50 + 50*sto_i

    #データの読み込み
    data_MCH = pl.read_csv("output_Folder/TPGshipdata/change_PGoutput/TPGship_model_and_result_MCH_capaMCH=" + str(storage) + "GWh.csv")
    data_loss = pl.read_csv("output_Folder/TPGshipdata/change_PGoutput/TPGship_model_and_result_Losselect_capaMCH=" + str(storage) + "GWh.csv")


    # 2019年から2023年までの年平均のMCH生成量を計算
    # MCH
    y_2019_MCH = data_MCH["2019 Total MCH[GWh]"]
    y_2020_MCH = data_MCH["2020 Total MCH[GWh]"]
    y_2021_MCH = data_MCH["2021 Total MCH[GWh]"]
    y_2022_MCH = data_MCH["2022 Total MCH[GWh]"]
    y_2023_MCH = data_MCH["2023 Total MCH[GWh]"]


    # Loss
    y_2019_loss = data_loss["2019 Total Losselect[GWh]"]
    y_2020_loss = data_loss["2020 Total Losselect[GWh]"]
    y_2021_loss = data_loss["2021 Total Losselect[GWh]"]
    y_2022_loss = data_loss["2022 Total Losselect[GWh]"]
    y_2023_loss = data_loss["2023 Total Losselect[GWh]"]


    # storageの数値で場合わけしてリストに各Storageの数値での年平均のMCH生成量を格納
    if storage == 50:
        for i in range(len(y_2019_MCH)):
            y_ave_MCH_sto50.append((y_2019_MCH[i] + y_2020_MCH[i] + y_2021_MCH[i] + y_2022_MCH[i] + y_2023_MCH[i])/5)
            y_ave_loss_sto50.append((y_2019_loss[i] + y_2020_loss[i] + y_2021_loss[i] + y_2022_loss[i] + y_2023_loss[i])/5)
    elif storage == 100:
        for i in range(len(y_2019_MCH)):
            y_ave_MCH_sto100.append((y_2019_MCH[i] + y_2020_MCH[i] + y_2021_MCH[i] + y_2022_MCH[i] + y_2023_MCH[i])/5)
            y_ave_loss_sto100.append((y_2019_loss[i] + y_2020_loss[i] + y_2021_loss[i] + y_2022_loss[i] + y_2023_loss[i])/5)
    elif storage == 150:
        for i in range(len(y_2019_MCH)):
            y_ave_MCH_sto150.append((y_2019_MCH[i] + y_2020_MCH[i] + y_2021_MCH[i] + y_2022_MCH[i] + y_2023_MCH[i])/5)
            y_ave_loss_sto150.append((y_2019_loss[i] + y_2020_loss[i] + y_2021_loss[i] + y_2022_loss[i] + y_2023_loss[i])/5)
    elif storage == 200:
        for i in range(len(y_2019_MCH)):
            y_ave_MCH_sto200.append((y_2019_MCH[i] + y_2020_MCH[i] + y_2021_MCH[i] + y_2022_MCH[i] + y_2023_MCH[i])/5)
            y_ave_loss_sto200.append((y_2019_loss[i] + y_2020_loss[i] + y_2021_loss[i] + y_2022_loss[i] + y_2023_loss[i])/5)
    elif storage == 250:
        for i in range(len(y_2019_MCH)):
            y_ave_MCH_sto250.append((y_2019_MCH[i] + y_2020_MCH[i] + y_2021_MCH[i] + y_2022_MCH[i] + y_2023_MCH[i])/5)
            y_ave_loss_sto250.append((y_2019_loss[i] + y_2020_loss[i] + y_2021_loss[i] + y_2022_loss[i] + y_2023_loss[i])/5)
    elif storage == 300:
        for i in range(len(y_2019_MCH)):
            y_ave_MCH_sto300.append((y_2019_MCH[i] + y_2020_MCH[i] + y_2021_MCH[i] + y_2022_MCH[i] + y_2023_MCH[i])/5)
            y_ave_loss_sto300.append((y_2019_loss[i] + y_2020_loss[i] + y_2021_loss[i] + y_2022_loss[i] + y_2023_loss[i])/5)
            
    
#データのプロット
#ここでスタイルを指定する
#今回は'science'と'ieee'を利用してみる
plt.style.use(['science','ieee','high-vis','grid'])

# 図の表示
fig = plt.figure()
ax = fig.add_subplot(1,1,1) #MCH
ax.set_xlim(xmin=0,xmax=0.3) #x軸の範囲を指定
ax.set_ylim(ymin=0,ymax=350) #y軸の範囲を指定
ax.plot(x,y_ave_MCH_sto50,label = '50GWh') #storage=50GWhの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_MCH_sto100,label = '100GWh') #storage=100GWhの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_MCH_sto150,label = '150GWh') #storage=150GWhの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_MCH_sto200,label = '200GWh') #storage=200GWhの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_MCH_sto250,label = '250GWh') #storage=250GWhの場合の年平均MCH生成量を表示
#ax.plot(x,y_ave_MCH_sto300,label = '300GWh') #storage=300GWhの場合の年平均MCH生成量を表示

ax.set_title('Average Annual MCH production')
#凡例はサイズを小さめにして表示
ax.legend(title = 'Q_capacityMCH',loc = 'lower right',fancybox = False, frameon = True, ncol = 2, fontsize = 4) #凡例の細かい設定
ax.set(xlabel = 'PG Output[GWh]') #x軸のラベル
ax.set(ylabel = 'Annual MCH production[GWh]') #y軸のラベル

fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1) #Loss
ax1.set_xlim(xmin=0,xmax=0.3) #x軸の範囲を指定
ax1.set_ylim(ymin=0,ymax=50) #y軸の範囲を指定
ax1.plot(x,y_ave_loss_sto50,label = '50GWh') #storage=50GWhの場合の年平均MCH生成量を表示
ax1.plot(x,y_ave_loss_sto100,label = '100GWh') #storage=100GWhの場合の年平均MCH生成量を表示
ax1.plot(x,y_ave_loss_sto150,label = '150GWh') #storage=150GWhの場合の年平均MCH生成量を表示
ax1.plot(x,y_ave_loss_sto200,label = '200GWh') #storage=200GWhの場合の年平均MCH生成量を表示
ax1.plot(x,y_ave_loss_sto250,label = '250GWh') #storage=250GWhの場合の年平均MCH生成量を表示
#ax1.plot(x,y_ave_loss_sto300,label = '300GWh') #storage=300GWhの場合の年平均MCH生成量を表示

ax1.set_title('Average Annual Loss')
#凡例はサイズを小さめにして表示
ax1.legend(title = 'Q_capacityMCH',loc = 'upper right',fancybox = False, frameon = True, ncol = 2, fontsize = 4) #凡例の細かい設定
ax1.set(xlabel = 'PG Output[GWh]') #x軸のラベル
ax1.set(ylabel = 'Annual Loss[GWh]') #y軸のラベル

plt.show()




```

```python
# 発電機格納タイプと非格納タイプのe_PG = 0.3~0.5の時の年平均MCH生成量をそれぞれのリストに格納
y_ave_genelift_e03 = []
y_ave_genelift_e04 = []
y_ave_genelift_e05 = []

y_ave_nolift_e03 = []
y_ave_nolift_e04 = []
y_ave_nolift_e05 = []

e_PG = 0.3
data_genelift = pl.read_csv("output_Folder/TPGshipdata/change_e_PG_LiftGenerator/TPGship_model_and_result_Losselect_ePG=" + str(e_PG) + "_and_Liftgene.csv")
x = data_genelift["Storage[GWh]"]

for e_i in range(3):
    e_PG = 0.3 + 0.1*e_i

    #データの読み込み
    data_genelift = pl.read_csv("output_Folder/TPGshipdata/change_e_PG_LiftGenerator/TPGship_model_and_result_Losselect_ePG=" + str(e_PG) + "_and_Liftgene.csv")
    data_nolift = pl.read_csv("output_Folder/TPGshipdata/change_e_PG/TPGship_model_and_result_Losselect_ePG=" + str(e_PG) + ".csv")

    # 2017年から2022年までの年平均のMCH生成量を計算
    # 発電機格納タイプの場合
    y_2017_genelift = data_genelift["2017 Total Loss[GWh]"]
    y_2018_genelift = data_genelift["2018 Total Loss[GWh]"]
    y_2019_genelift = data_genelift["2019 Total Loss[GWh]"]
    y_2020_genelift = data_genelift["2020 Total Loss[GWh]"]
    y_2021_genelift = data_genelift["2021 Total Loss[GWh]"]
    y_2022_genelift = data_genelift["2022 Total Loss[GWh]"]

    # 発電機非格納タイプの場合
    y_2017_nolift = data_nolift["2017 Total Loss[GWh]"]
    y_2018_nolift = data_nolift["2018 Total Loss[GWh]"]
    y_2019_nolift = data_nolift["2019 Total Loss[GWh]"]
    y_2020_nolift = data_nolift["2020 Total Loss[GWh]"]
    y_2021_nolift = data_nolift["2021 Total Loss[GWh]"]
    y_2022_nolift = data_nolift["2022 Total Loss[GWh]"]    

    # e_PGの数値で場合わけしてリストに各Storageの数値での年平均のMCH生成量を格納
    if e_PG == 0.3:
        for i in range(len(y_2017_genelift)):
            y_ave_genelift_e03.append((y_2017_genelift[i] + y_2018_genelift[i] + y_2019_genelift[i] + y_2020_genelift[i] + y_2021_genelift[i] + y_2022_genelift[i])/6)
            y_ave_nolift_e03.append((y_2017_nolift[i] + y_2018_nolift[i] + y_2019_nolift[i] + y_2020_nolift[i] + y_2021_nolift[i] + y_2022_nolift[i])/6)
    elif e_PG == 0.4:
        for i in range(len(y_2017_genelift)):
            y_ave_genelift_e04.append((y_2017_genelift[i] + y_2018_genelift[i] + y_2019_genelift[i] + y_2020_genelift[i] + y_2021_genelift[i] + y_2022_genelift[i])/6)
            y_ave_nolift_e04.append((y_2017_nolift[i] + y_2018_nolift[i] + y_2019_nolift[i] + y_2020_nolift[i] + y_2021_nolift[i] + y_2022_nolift[i])/6)
    elif e_PG == 0.5:
        for i in range(len(y_2017_genelift)):
            y_ave_genelift_e05.append((y_2017_genelift[i] + y_2018_genelift[i] + y_2019_genelift[i] + y_2020_genelift[i] + y_2021_genelift[i] + y_2022_genelift[i])/6)
            y_ave_nolift_e05.append((y_2017_nolift[i] + y_2018_nolift[i] + y_2019_nolift[i] + y_2020_nolift[i] + y_2021_nolift[i] + y_2022_nolift[i])/6)
    
    
#データのプロット
#ここでスタイルを指定する
#今回は'science'と'ieee'を利用してみる
plt.style.use(['science','ieee','high-vis','grid'])

# 図の表示
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_xlim(xmin=0,xmax=300) #x軸の範囲を指定
ax.set_ylim(ymin=0,ymax=40) #y軸の範囲を指定
ax.plot(x,y_ave_genelift_e03,label = 'Storable Generators(e_PG=0.3)') #e_PG=0.3の発電機格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_nolift_e03,label = 'Non-Storable Generators(e_PG=0.3)') #e_PG=0.3の発電機非格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_genelift_e04,label = 'Storable Generators(e_PG=0.4)') #e_PG=0.4の発電機格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_nolift_e04,label = 'Non-Storable Generators(e_PG=0.4)') #e_PG=0.4の発電機非格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_genelift_e05,label = 'Storable Generators(e_PG=0.5)') #e_PG=0.5の発電機格納タイプの場合の年平均MCH生成量を表示
ax.plot(x,y_ave_nolift_e05,label = 'Non-Storable Generators(e_PG=0.5)') #e_PG=0.5の発電機非格納タイプの場合の年平均MCH生成量を表示
ax.set_title('Average Annual power consumption')
#凡例はサイズを小さめにして表示
ax.legend(title = 'Type of Generator',loc = 'lower right',fancybox = False, frameon = True, ncol = 2, fontsize = 4) #凡例の細かい設定

ax.set(xlabel = 'Q_capacityMCH[GWh]') #x軸のラベル
ax.set(ylabel = 'Annual Power Consumption[GWh]') #y軸のラベル
plt.show()
```
